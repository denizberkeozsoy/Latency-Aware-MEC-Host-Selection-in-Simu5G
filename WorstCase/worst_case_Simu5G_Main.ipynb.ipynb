```python
def summarize_basic_stats(before_df, after_df, before_path='[Before File]', after_path='[After File]'):
    import pandas as pd
    from IPython.display import display

    def clean_describe(df, label, path):
        print(f"===== {label} Dataset Summary =====")
        print(f"File Path: {path}")
        print("Shape:", df.shape)
        print("Columns:", list(df.columns))
        print("\nData Types:\n", df.dtypes.value_counts())

        # Numeric summary
        print("\n--- Numeric Summary ---")
        display(df.describe(include='number'))

        # Categorical summary
        print("\n--- Categorical Summary ---")
        display(df.describe(include='object'))

    print("\n\n===== BEFORE Dataset Summary =====")
    before_copy = before_df.copy()
    before_copy['value'] = pd.to_numeric(before_copy['value'], errors='coerce')
    clean_describe(before_copy, "Before", before_path)

    print("\n\n===== AFTER Dataset Summary =====")
    after_copy = after_df.copy()
    after_copy['value'] = pd.to_numeric(after_copy['value'], errors='coerce')
    clean_describe(after_copy, "After", after_path)
```

```python
def compare_scalar_metric_counts(before_path, after_path):
    import pandas as pd

    before_df = pd.read_csv(before_path, low_memory=False)
    after_df = pd.read_csv(after_path, low_memory=False)

    before_scalar = before_df[before_df['type'] == 'scalar']
    after_scalar = after_df[after_df['type'] == 'scalar']

    before_counts = before_scalar['name'].value_counts()
    after_counts = after_scalar['name'].value_counts()

    all_metrics = list(set(before_counts.index) | set(after_counts.index))

    summary = pd.DataFrame(index=all_metrics)
    summary['Before Count'] = before_counts
    summary['After Count'] = after_counts
    summary = summary.fillna(0).astype(int)
    summary['Difference'] = summary['After Count'] - summary['Before Count']

    print("===== Metric Count Comparison =====")
    print(summary.sort_values(by='Difference', ascending=False).head(15))
```

```python
def compare_scalar_metric_counts_visually(before_path, after_path, save_dir=r"C:\Users\deniz\OneDrive\Masaüstü\research_evaluation_photos"):
    import pandas as pd
    import matplotlib.pyplot as plt
    import os
    from pathlib import Path

    before = pd.read_csv(before_path, low_memory=False)
    after = pd.read_csv(after_path, low_memory=False)

    before_scalar = before[before['type'] == 'scalar']
    after_scalar = after[after['type'] == 'scalar']

    before_counts = before_scalar['name'].value_counts()
    after_counts = after_scalar['name'].value_counts()

    combined = pd.DataFrame({
        'Before': before_counts,
        'After': after_counts
    }).fillna(0)

    combined['Change'] = combined['After'] - combined['Before']
    top_changed = combined.sort_values(by='Change', ascending=False).head(10)

    # Prepare plot folder and filename
    def get_next_plot_index(directory, prefix="count_diff_plot_"):
        Path(directory).mkdir(parents=True, exist_ok=True)
        existing = [f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith(".png")]
        indices = [int(f.split("_")[3].split(".")[0]) for f in existing if f.split("_")[3].split(".")[0].isdigit()]
        return max(indices, default=0) + 1

    plot_idx = get_next_plot_index(save_dir)

    # Plot
    top_changed[['Before', 'After']].plot(kind='bar', figsize=(10, 5), color=['steelblue', 'salmon'])
    plt.title("Top 10 Changed Scalar Metric Counts")
    plt.ylabel("Metric Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Save
    save_path = os.path.join(save_dir, f"count_diff_plot_{plot_idx}.png")
    plt.savefig(save_path)
    plt.show()
```

```python
def analyze_new_metrics_boxplot(before_path, after_path, save_dir=r"C:\Users\deniz\OneDrive\Masaüstü\research_evaluation_photos"):
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os
    from pathlib import Path

    metrics = {
        'incomingPacketLengths:sum': 'Incoming Packet Lengths',
        'outgoingPacketLengths:sum': 'Outgoing Packet Lengths',
        'rlcCellThroughputUl:mean': 'RLC Throughput UL',
        'macCellThroughputDl:mean': 'MAC Throughput DL',
        'txPk:sum(packetBytes)': 'Bytes Transmitted',
        'rxPkOk:sum(packetBytes)': 'Bytes Received'
    }

    before = pd.read_csv(before_path, low_memory=False)
    after = pd.read_csv(after_path, low_memory=False)

    before_scalar = before[before['type'] == 'scalar']
    after_scalar = after[after['type'] == 'scalar']

    before_metrics = before_scalar[before_scalar['name'].isin(metrics.keys())].copy()
    after_metrics = after_scalar[after_scalar['name'].isin(metrics.keys())].copy()

    before_metrics['value'] = pd.to_numeric(before_metrics['value'], errors='coerce')
    after_metrics['value'] = pd.to_numeric(after_metrics['value'], errors='coerce')

    before_metrics['phase'] = 'Before'
    after_metrics['phase'] = 'After'

    combined = pd.concat([before_metrics, after_metrics])
    combined['Metric'] = combined['name'].map(metrics)

    sns.set(style="whitegrid")

    # Ensure save directory exists
    Path(save_dir).mkdir(parents=True, exist_ok=True)

    # Determine next index
    def get_next_plot_index(directory, prefix="metric_boxplot_"):
        existing = [f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith(".png")]
        indices = [int(f.split("_")[-1].split(".")[0]) for f in existing if f.split("_")[-1].split(".")[0].isdigit()]
        return max(indices, default=0) + 1

    plot_idx = get_next_plot_index(save_dir)

    for metric in metrics.values():
        plt.figure(figsize=(10, 5))
        subset = combined[combined['Metric'] == metric]
        sns.boxplot(data=subset, x='phase', y='value')  # No color change
        plt.title(f"{metric}: Before vs After")
        plt.tight_layout()

        # Save and show
        save_path = os.path.join(save_dir, f"metric_boxplot_{plot_idx}.png")
        plt.savefig(save_path)
        plt.show()
        plot_idx += 1
```

```python
def visualize_metric_correlations(before_df, after_df, save_dir=r"C:\Users\deniz\OneDrive\Masaüstü\research_evaluation_photos"):
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os
    from pathlib import Path

    # Ensure 'value' is numeric
    before_df['value'] = pd.to_numeric(before_df['value'], errors='coerce')
    after_df['value'] = pd.to_numeric(after_df['value'], errors='coerce')

    # Filter scalar-only data
    before_clean = before_df[before_df['type'] == 'scalar'].copy()
    after_clean = after_df[after_df['type'] == 'scalar'].copy()

    # Pivot tables
    before_pivot = before_clean.pivot_table(index='module', columns='name', values='value', aggfunc='mean')
    after_pivot = after_clean.pivot_table(index='module', columns='name', values='value', aggfunc='mean')

    before_corr = before_pivot.corr().dropna(how='all', axis=1).dropna(how='all', axis=0)
    after_corr = after_pivot.corr().dropna(how='all', axis=1).dropna(how='all', axis=0)

    # Create save directory
    Path(save_dir).mkdir(parents=True, exist_ok=True)

    # Determine next index
    def next_index(prefix):
        files = [f for f in os.listdir(save_dir) if f.startswith(prefix) and f.endswith('.png')]
        indices = [int(f.split('_')[-2]) for f in files if f.split('_')[-2].isdigit()]
        return max(indices, default=0) + 1

    idx = next_index("correlation_heatmap")

    # Plot and save BEFORE heatmap
    plt.figure(figsize=(12, 6))
    sns.heatmap(before_corr, cmap='coolwarm', center=0, annot=False, linewidths=0.5)
    plt.title("Correlation Matrix of Metrics (Before)")
    plt.tight_layout()
    before_path = os.path.join(save_dir, f"correlation_heatmap_{idx}_before.png")
    plt.savefig(before_path)
    plt.show()

    # Plot and save AFTER heatmap
    plt.figure(figsize=(12, 6))
    sns.heatmap(after_corr, cmap='coolwarm', center=0, annot=False, linewidths=0.5)
    plt.title("Correlation Matrix of Metrics (After)")
    plt.tight_layout()
    after_path = os.path.join(save_dir, f"correlation_heatmap_{idx}_after.png")
    plt.savefig(after_path)
    plt.show()

    # Plot and save DIFF heatmap
    shared_cols = before_corr.columns.intersection(after_corr.columns)
    shared_rows = before_corr.index.intersection(after_corr.index)
    corr_diff = after_corr.loc[shared_rows, shared_cols] - before_corr.loc[shared_rows, shared_cols]

    plt.figure(figsize=(12, 6))
    sns.heatmap(corr_diff, cmap='bwr', center=0, annot=False, linewidths=0.5)
    plt.title("Correlation Difference Matrix (After - Before)")
    plt.tight_layout()
    diff_path = os.path.join(save_dir, f"correlation_heatmap_{idx}_diff.png")
    plt.savefig(diff_path)
    plt.show()
```

```python
def compare_metric_means_and_stds(before_df, after_df, top_n=10, sort_by='Delta Mean',
                                   save_dir=r"C:\Users\deniz\OneDrive\Masaüstü\research_evaluation_photos"):
    import pandas as pd
    import matplotlib.pyplot as plt
    import os
    from pathlib import Path

    # Ensure numeric
    before_df['value'] = pd.to_numeric(before_df['value'], errors='coerce')
    after_df['value'] = pd.to_numeric(after_df['value'], errors='coerce')

    # Filter scalar
    before_clean = before_df[before_df['type'] == 'scalar'].dropna(subset=['name', 'value'])
    after_clean = after_df[after_df['type'] == 'scalar'].dropna(subset=['name', 'value'])

    common_metrics = set(before_clean['name']).intersection(set(after_clean['name']))
    summary_stats = []

    for metric in common_metrics:
        b_vals = before_clean[before_clean['name'] == metric]['value']
        a_vals = after_clean[after_clean['name'] == metric]['value']

        if not b_vals.empty and not a_vals.empty:
            summary_stats.append({
                "Metric": metric,
                "Before Mean": b_vals.mean(),
                "After Mean": a_vals.mean(),
                "Before Std": b_vals.std(),
                "After Std": a_vals.std(),
                "Delta Mean": a_vals.mean() - b_vals.mean(),
                "Abs Delta Mean": abs(a_vals.mean() - b_vals.mean())
            })

    summary_df = pd.DataFrame(summary_stats)

    if sort_by not in summary_df.columns:
        sort_by = "Delta Mean"

    summary_df_sorted = summary_df.sort_values(sort_by, ascending=False).reset_index(drop=True)

    print(f"\n===== Top {top_n} Metrics Sorted by {sort_by} =====\n")
    print(summary_df_sorted[['Metric', 'Before Mean', 'After Mean', 'Delta Mean', 'Before Std', 'After Std']].head(top_n))

    # Visualization
    top_metrics = summary_df_sorted.head(top_n)
    labels = top_metrics['Metric']
    before_means = top_metrics['Before Mean']
    after_means = top_metrics['After Mean']

    plt.figure(figsize=(12, 6))
    x = range(len(labels))
    plt.bar(x, before_means, width=0.4, label='Before', align='center')
    plt.bar([i + 0.4 for i in x], after_means, width=0.4, label='After', align='center')
    plt.xticks([i + 0.2 for i in x], labels, rotation=45, ha='right')
    plt.title(f"Top {top_n} Metrics - Mean Comparison")
    plt.ylabel("Mean Value")
    plt.legend()
    plt.tight_layout()

    # Save image to desktop folder
    Path(save_dir).mkdir(parents=True, exist_ok=True)
    existing_files = [f for f in os.listdir(save_dir) if f.startswith("metric_mean_comparison_") and f.endswith(".png")]
    indices = [int(f.split('_')[-1].split('.')[0]) for f in existing_files if f.split('_')[-1].split('.')[0].isdigit()]
    next_idx = max(indices, default=0) + 1
    filename = os.path.join(save_dir, f"metric_mean_comparison_{next_idx}.png")
    plt.savefig(filename)

    plt.show()

    return summary_df_sorted
```

```python
# Load your CSVs first
before_df = pd.read_csv("Simu5GBefore.csv", low_memory=False)
after_df = pd.read_csv("Simu5GAfter.csv", low_memory=False)

# Run comparison
compare_metric_means_and_stds(before_df, after_df, top_n=15, sort_by='Abs Delta Mean')
```

```python
def compare_queue_metrics(before_df, after_df, metric_keywords=None):
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os
    from pathlib import Path

    if metric_keywords is None:
        metric_keywords = ['queueLength', 'queueingDelay', 'droppedPackets', 'queueSize']

    # Directory to save images
    save_dir = r"C:\Users\deniz\OneDrive\Masaüstü\research_evaluation_photos"
    Path(save_dir).mkdir(parents=True, exist_ok=True)

    # Find current max index
    existing = [f for f in os.listdir(save_dir) if f.startswith("queue_metric_") and f.endswith(".png")]
    indices = []
    for f in existing:
        try:
            idx_part = f.split('_')[2]
            if idx_part.isdigit():
                indices.append(int(idx_part))
        except:
            continue
    next_index = max(indices, default=0) + 1

    # Filter and clean scalar data
    before_df = before_df[before_df['type'] == 'scalar'].copy()
    after_df = after_df[after_df['type'] == 'scalar'].copy()
    before_df['value'] = pd.to_numeric(before_df['value'], errors='coerce')
    after_df['value'] = pd.to_numeric(after_df['value'], errors='coerce')

    for keyword in metric_keywords:
        before_match = before_df[before_df['name'].str.contains(keyword, case=False, na=False)]
        after_match = after_df[after_df['name'].str.contains(keyword, case=False, na=False)]

        if before_match.empty or after_match.empty:
            print(f"⚠️ No data found for keyword: '{keyword}' — skipping.")
            continue

        combined = pd.concat([
            before_match.assign(phase='Before'),
            after_match.assign(phase='After')
        ])

        print(f"\n===== Summary for Queue Metric: {keyword} =====")
        display(combined.groupby('phase')['value'].describe())

        # Plot
        plt.figure(figsize=(10, 5))
        sns.boxplot(data=combined, x='phase', y='value', hue='phase', palette='Set2', legend=False)
        plt.title(f"Queue Metric Comparison: {keyword}")
        plt.ylabel("Value")
        plt.tight_layout()

        # Save to desktop
        filename = f"queue_metric_{next_index}_{keyword}.png"
        filepath = os.path.join(save_dir, filename)
        plt.savefig(filepath, dpi=300)
        print(f"📸 Saved to: {filepath}")

        plt.show()
        next_index += 1
```

```python
compare_queue_metrics(before_df, after_df, metric_keywords=['queueBitLength', 'queueDelay', 'queueLength'])
```
