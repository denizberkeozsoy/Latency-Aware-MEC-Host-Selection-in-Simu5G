```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option("display.max_columns", None)
sns.set(style="whitegrid", palette="muted")
```

```python
def summarize_basic_stats_all(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
):
    import pandas as pd
    from IPython.display import display

    def clean_describe(df, label, path):
        print(f"===== {label} Dataset Summary =====")
        print(f"File Path: {path}")
        print("Shape:", df.shape)
        print("Columns:", list(df.columns))
        print("\nData Types:\n", df.dtypes.value_counts())
        print("\n--- Numeric Summary ---")
        display(df.describe(include='number'))
        print("\n--- Categorical Summary ---")
        display(df.describe(include='object'))

    def prepare(df):
        df = df.copy()
        if 'value' in df.columns:
            df['value'] = pd.to_numeric(df['value'], errors='coerce')
        return df

    before_df = pd.read_csv(before_path, low_memory=False)
    worst_df = pd.read_csv(worst_path, low_memory=False)
    moderate_df = pd.read_csv(moderate_path, low_memory=False)
    best_df = pd.read_csv(best_path, low_memory=False)

    print("\n\n===== BEFORE Dataset Summary =====")
    clean_describe(prepare(before_df), "Before", before_path)

    print("\n\n===== AFTER (WorstCase) Dataset Summary =====")
    clean_describe(prepare(worst_df), "After - WorstCase", worst_path)

    print("\n\n===== AFTER (ModerateCase) Dataset Summary =====")
    clean_describe(prepare(moderate_df), "After - ModerateCase", moderate_path)

    print("\n\n===== AFTER (BestCase) Dataset Summary =====")
    clean_describe(prepare(best_df), "After - BestCase", best_path)
```

```python
def summarize_simulation_data_visually(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
    top_scalar_names=10,
    bins=50
):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt

    # ---------- Load & normalize ----------
    def load_df(path):
        df = pd.read_csv(path, low_memory=False)
        if 'value' in df.columns:
            df['value'] = pd.to_numeric(df['value'], errors='coerce')
        return df

    datasets = {
        "BEFORE": load_df(before_path),
        "AFTER_WorstCase": load_df(worst_path),
        "AFTER_ModerateCase": load_df(moderate_path),
        "AFTER_BestCase": load_df(best_path),
    }

    # Distinct colors for clarity
    colors = {
        "BEFORE": "#1f77b4",            # blue
        "AFTER_WorstCase": "#d62728",   # red
        "AFTER_ModerateCase": "#ff7f0e",# orange
        "AFTER_BestCase": "#2ca02c",    # green
    }

    def has_cols(df, cols):
        return all(c in df.columns for c in cols)

    # ---------- 0) Quick console summary ----------
    for label, df in datasets.items():
        print(f"\n===== Summary for: {label} =====")
        print("Shape:", df.shape)
        print("Columns:", list(df.columns))

    # ---------- 1) Overlaid histogram of overall 'value' ----------
    have_any_values = False
    plt.figure(figsize=(8, 5))
    for label, df in datasets.items():
        if 'value' in df.columns:
            vals = df['value'].dropna()
            if not vals.empty:
                have_any_values = True
                plt.hist(vals, bins=bins, alpha=0.35, label=label, color=colors[label], density=False)
    if have_any_values:
        plt.title("Overall Value Distribution (All Records)")
        plt.xlabel("value")
        plt.ylabel("frequency")
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        plt.close()
        print("[Info] Skipped overall value histogram: no numeric 'value' found in any dataset.")

    # ---------- 2) Boxplot of 'value' for scalar-only rows ----------
    scalar_values = []
    labels_for_box = []
    for label, df in datasets.items():
        if has_cols(df, ['type', 'value']):
            vals = df.loc[df['type'] == 'scalar', 'value'].dropna()
            if not vals.empty:
                scalar_values.append(vals.values)
                labels_for_box.append(label)

    if scalar_values:
        plt.figure(figsize=(8, 5))
        bp = plt.boxplot(scalar_values, labels=labels_for_box, patch_artist=True)
        for patch, label in zip(bp['boxes'], labels_for_box):
            patch.set_facecolor(colors[label])
            patch.set_alpha(0.5)
        plt.title("Scalar-Only Value Distribution (Boxplot)")
        plt.ylabel("value")
        plt.tight_layout()
        plt.show()
    else:
        print("[Info] Skipped scalar boxplot: no scalar numeric 'value' in any dataset.")

    # ---------- 3) Grouped bar of 'type' counts ----------
    type_set = set()
    type_counts = {}
    for label, df in datasets.items():
        if 'type' in df.columns:
            counts = df['type'].value_counts()
            type_counts[label] = counts
            type_set.update(counts.index)

    if type_set:
        types_sorted = sorted(type_set)
        x = np.arange(len(types_sorted))
        width = 0.2

        plt.figure(figsize=(10, 5))
        offsets = {
            "BEFORE": -1.5*width,
            "AFTER_WorstCase": -0.5*width,
            "AFTER_ModerateCase": 0.5*width,
            "AFTER_BestCase": 1.5*width,
        }

        for label in datasets.keys():
            counts = type_counts.get(label, pd.Series(dtype=int))
            y = [int(counts.get(t, 0)) for t in types_sorted]
            plt.bar(x + offsets[label], y, width, label=label, color=colors[label])

        plt.xticks(x, types_sorted, rotation=20)
        plt.ylabel("count")
        plt.title("Record Type Counts (Grouped)")
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print("[Info] Skipped type grouped bar: no 'type' column found in any dataset.")

    # ---------- 4) Grouped bar for top scalar metric names ----------
    name_freq_overall = pd.Series(dtype=int)
    scalar_name_counts = {}
    any_scalar_names = False

    for label, df in datasets.items():
        if has_cols(df, ['type', 'name']):
            scalars = df.loc[df['type'] == 'scalar', 'name'].dropna()
            vc = scalars.value_counts()
            if not vc.empty:
                any_scalar_names = True
                scalar_name_counts[label] = vc
                name_freq_overall = name_freq_overall.add(vc, fill_value=0)

    if any_scalar_names and not name_freq_overall.empty:
        top_names = list(name_freq_overall.sort_values(ascending=False).head(top_scalar_names).index)
        x = np.arange(len(top_names))
        width = 0.2

        plt.figure(figsize=(max(10, len(top_names)*0.8), 5))
        offsets = {
            "BEFORE": -1.5*width,
            "AFTER_WorstCase": -0.5*width,
            "AFTER_ModerateCase": 0.5*width,
            "AFTER_BestCase": 1.5*width,
        }

        for label in datasets.keys():
            vc = scalar_name_counts.get(label, pd.Series(dtype=int))
            y = [int(vc.get(n, 0)) for n in top_names]
            plt.bar(x + offsets[label], y, width, label=label, color=colors[label])

        plt.xticks(x, top_names, rotation=30, ha='right')
        plt.ylabel("count")
        plt.title(f"Top {top_scalar_names} Scalar Metric Names (Counts)")
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print("[Info] Skipped top scalar names grouped bar: no scalar 'name' data available.")
```

```python
def describe_scalar_metrics_visually(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
    top_scalar_names=10
):
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns

    # ---- Load & prep helper ----
    def load_scalar_df(path):
        df = pd.read_csv(path, low_memory=False)
        if 'type' not in df.columns:
            return pd.DataFrame(columns=['name', 'value'])
        df = df[df['type'] == 'scalar'].copy()
        if 'name' not in df.columns:
            df['name'] = np.nan
        if 'value' in df.columns:
            df['value'] = pd.to_numeric(df['value'], errors='coerce')
        else:
            df['value'] = np.nan
        return df

    datasets = {
        "BEFORE": load_scalar_df(before_path),
        "AFTER_WorstCase": load_scalar_df(worst_path),
        "AFTER_ModerateCase": load_scalar_df(moderate_path),
        "AFTER_BestCase": load_scalar_df(best_path),
    }

    # console summary (keeps original spirit)
    for label, sdf in datasets.items():
        print(f"\n===== Scalar Metrics Summary for: {label} =====")
        print("Total scalar entries:", len(sdf))
        print("Unique scalar metrics:", sdf['name'].nunique())

    # distinct colors per dataset
    colors = {
        "BEFORE": "#1f77b4",             # blue
        "AFTER_WorstCase": "#d62728",    # red
        "AFTER_ModerateCase": "#ff7f0e", # orange
        "AFTER_BestCase": "#2ca02c",     # green
    }

    # ---------- Plot 1: Grouped bar of Top-N scalar metric names (by frequency overall) ----------
    name_freq_overall = pd.Series(dtype=int)
    per_ds_counts = {}
    for label, sdf in datasets.items():
        vc = sdf['name'].dropna().value_counts()
        per_ds_counts[label] = vc
        name_freq_overall = name_freq_overall.add(vc, fill_value=0)

    if not name_freq_overall.empty:
        top_names = list(name_freq_overall.sort_values(ascending=False).head(top_scalar_names).index)
        x = np.arange(len(top_names))
        width = 0.2
        offsets = {
            "BEFORE": -1.5*width,
            "AFTER_WorstCase": -0.5*width,
            "AFTER_ModerateCase": 0.5*width,
            "AFTER_BestCase": 1.5*width,
        }

        plt.figure(figsize=(max(10, len(top_names)*0.85), 5))
        for label in datasets.keys():
            vc = per_ds_counts.get(label, pd.Series(dtype=int))
            y = [int(vc.get(n, 0)) for n in top_names]
            plt.bar(x + offsets[label], y, width, label=label, color=colors[label])
        plt.xticks(x, top_names, rotation=30, ha='right')
        plt.ylabel("Count")
        plt.title(f"Top {top_scalar_names} Scalar Metric Names  Counts by Dataset")
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print("[Info] Skipped Top-N names plot: no scalar 'name' data found.")

    # ---------- Plot 2: Boxplot of scalar values per dataset ----------
    value_series = []
    labels = []
    for label, sdf in datasets.items():
        vals = sdf['value'].dropna()
        if not vals.empty:
            value_series.append(vals.values)
            labels.append(label)  # <-- fixed typo here

    if value_series:
        plt.figure(figsize=(8, 5))
        bp = plt.boxplot(value_series, labels=labels, patch_artist=True)
        for patch, lbl in zip(bp['boxes'], labels):
            patch.set_facecolor(colors[lbl])
            patch.set_alpha(0.5)
        plt.title("Scalar Metric Values  Distribution by Dataset (Boxplot)")
        plt.ylabel("value")
        plt.tight_layout()
        plt.show()
    else:
        print("[Info] Skipped boxplot: no numeric scalar 'value' across datasets.")

    # ---------- Optional: KDE overlay for scalar values ----------
    any_kde = False
    plt.figure(figsize=(8, 5))
    for label, sdf in datasets.items():
        vals = sdf['value'].dropna()
        if not vals.empty:
            any_kde = True
            sns.kdeplot(vals, label=label, color=colors[label], fill=True, alpha=0.25)
    if any_kde:
        plt.title("Scalar Metric Values  KDE Overlay")
        plt.xlabel("value")
        plt.ylabel("density")
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        plt.close()
        print("[Info] Skipped KDE: no numeric scalar 'value' across datasets.")
```

```python
def compare_scalar_metric_counts_visually(before_path, after_path, save_dir=r"C:\Users\deniz\OneDrive\Masast\research_evaluation_photos"):
    import pandas as pd
    import matplotlib.pyplot as plt
    import os
    from pathlib import Path

    before = pd.read_csv(before_path, low_memory=False)
    after = pd.read_csv(after_path, low_memory=False)

    before_scalar = before[before['type'] == 'scalar']
    after_scalar = after[after['type'] == 'scalar']

    before_counts = before_scalar['name'].value_counts()
    after_counts = after_scalar['name'].value_counts()

    combined = pd.DataFrame({
        'Before': before_counts,
        'After': after_counts
    }).fillna(0)

    combined['Change'] = combined['After'] - combined['Before']
    top_changed = combined.sort_values(by='Change', ascending=False).head(10)

    # Prepare plot folder and filename
    def get_next_plot_index(directory, prefix="count_diff_plot_"):
        Path(directory).mkdir(parents=True, exist_ok=True)
        existing = [f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith(".png")]
        indices = [int(f.split("_")[3].split(".")[0]) for f in existing if f.split("_")[3].split(".")[0].isdigit()]
        return max(indices, default=0) + 1

    plot_idx = get_next_plot_index(save_dir)

    # Plot
    top_changed[['Before', 'After']].plot(kind='bar', figsize=(10, 5), color=['steelblue', 'salmon'])
    plt.title("Top 10 Changed Scalar Metric Counts")
    plt.ylabel("Metric Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Save
    save_path = os.path.join(save_dir, f"count_diff_plot_{plot_idx}.png")
    plt.savefig(save_path)
    plt.show()
```

```python
def analyze_new_metrics_boxplot(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
    save_dir=r"C:\Users\deniz\OneDrive\Masast\research_evaluation_photos"
):
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os
    from pathlib import Path

    # Target scalar metrics (name -> pretty label)
    metrics = {
        'incomingPacketLengths:sum': 'Incoming Packet Lengths',
        'outgoingPacketLengths:sum': 'Outgoing Packet Lengths',
        'rlcCellThroughputUl:mean': 'RLC Throughput UL',
        'macCellThroughputDl:mean': 'MAC Throughput DL',
        'txPk:sum(packetBytes)': 'Bytes Transmitted',
        'rxPkOk:sum(packetBytes)': 'Bytes Received'
    }

    # --- Load CSVs ---
    def load_df(path):
        df = pd.read_csv(path, low_memory=False)
        return df

    before   = load_df(before_path)
    worst    = load_df(worst_path)
    moderate = load_df(moderate_path)
    best     = load_df(best_path)

    # --- Keep scalar rows & selected metrics, add phase labels ---
    def prep(df, phase):
        if 'type' not in df.columns:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf = df[df['type'] == 'scalar'].copy()
        if 'name' not in sdf.columns or 'value' not in sdf.columns:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf = sdf[sdf['name'].isin(metrics.keys())].copy()
        sdf['value'] = pd.to_numeric(sdf['value'], errors='coerce')
        sdf['phase'] = phase
        return sdf

    before_s   = prep(before,   'Before')
    worst_s    = prep(worst,    'After - Worst')
    moderate_s = prep(moderate, 'After - Moderate')
    best_s     = prep(best,     'After - Best')

    combined = pd.concat([before_s, worst_s, moderate_s, best_s], ignore_index=True)
    if combined.empty:
        print("[Info] No matching scalar metrics found in the provided files.")
        return

    # Map to pretty metric labels and order phases for consistent plots
    combined['Metric'] = combined['name'].map(metrics)
    combined['phase'] = pd.Categorical(
        combined['phase'],
        categories=['Before', 'After - Worst', 'After - Moderate', 'After - Best'],
        ordered=True
    )

    sns.set(style="whitegrid")

    # Ensure save directory exists & next index calculation (same behavior as your original)
    Path(save_dir).mkdir(parents=True, exist_ok=True)

    def get_next_plot_index(directory, prefix="metric_boxplot_"):
        existing = [f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith(".png")]
        indices = []
        for f in existing:
            stem = f.rsplit(".", 1)[0]
            try:
                indices.append(int(stem.split("_")[-1]))
            except Exception:
                pass
        return (max(indices) if indices else 0) + 1

    plot_idx = get_next_plot_index(save_dir)

    # --- One figure per metric: boxplot of value grouped by phase ---
    for pretty_name in metrics.values():
        subset = combined[(combined['Metric'] == pretty_name) & (~combined['value'].isna())]
        if subset.empty:
            print(f"[Info] Skipping '{pretty_name}': no numeric values found.")
            continue

        plt.figure(figsize=(10, 5))
        sns.boxplot(data=subset, x='phase', y='value')  # keep your no color change style
        plt.title(f"{pretty_name}: Before vs After Variants")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.tight_layout()

        save_path = os.path.join(save_dir, f"metric_boxplot_{plot_idx}.png")
        plt.savefig(save_path)
        plt.show()
        plot_idx += 1
```

```python
def analyze_new_metrics_meanline(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
    save_dir=r"C:\Users\deniz\OneDrive\Masast\research_evaluation_photos"
):
    import pandas as pd
    import matplotlib.pyplot as plt
    import os
    from pathlib import Path

    # (raw name) -> (pretty label)
    metrics = {
        'incomingPacketLengths:sum': 'Incoming Packet Lengths',
        'outgoingPacketLengths:sum': 'Outgoing Packet Lengths',
        'rlcCellThroughputUl:mean': 'RLC Throughput UL',
        'macCellThroughputDl:mean': 'MAC Throughput DL',
        'txPk:sum(packetBytes)': 'Bytes Transmitted',
        'rxPkOk:sum(packetBytes)': 'Bytes Received'
    }

    # desired display order
    metric_order = [
        'Bytes Received',
        'Bytes Transmitted',
        'Incoming Packet Lengths',
        'MAC Throughput DL',
        'Outgoing Packet Lengths',
        'RLC Throughput UL',
    ]

    # --- helpers ---
    def load_df(path):
        return pd.read_csv(path, low_memory=False)

    def prep(df, phase):
        if 'type' not in df.columns:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf = df[df['type'] == 'scalar'].copy()
        if 'name' not in sdf.columns or 'value' not in sdf.columns:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf = sdf[sdf['name'].isin(metrics.keys())].copy()
        if sdf.empty:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf['value'] = pd.to_numeric(sdf['value'], errors='coerce')
        sdf['phase'] = phase
        return sdf

    # --- load + prepare each dataset ---
    before   = prep(load_df(before_path),   'Before')
    worst    = prep(load_df(worst_path),    'After - Worst')
    moderate = prep(load_df(moderate_path), 'After - Moderate')
    best     = prep(load_df(best_path),     'After - Best')

    combined = pd.concat([before, worst, moderate, best], ignore_index=True)
    if combined.empty:
        print("[Info] No matching scalar metrics found in the provided files.")
        return

    # map to pretty names and order phases
    combined['Metric'] = combined['name'].map(metrics)
    combined = combined.dropna(subset=['Metric', 'value'])
    combined['phase'] = pd.Categorical(
        combined['phase'],
        categories=['Before', 'After - Worst', 'After - Moderate', 'After - Best'],
        ordered=True
    )

    # compute mean per metric per phase
    mean_df = (
        combined.groupby(['Metric', 'phase'])['value']
        .mean()
        .unstack(fill_value=float('nan'))
    )

    # enforce metric order (keep only those present)
    present_order = [m for m in metric_order if m in mean_df.index]
    if not present_order:
        print("[Info] None of the target metrics are present after filtering.")
        return
    mean_df = mean_df.loc[present_order]

    # ensure save dir & next index
    Path(save_dir).mkdir(parents=True, exist_ok=True)
    def get_next_plot_index(directory, prefix="metric_meanline_"):
        existing = [f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith(".png")]
        idxs = []
        for f in existing:
            try:
                idxs.append(int(f.rsplit(".", 1)[0].split("_")[-1]))
            except Exception:
                pass
        return (max(idxs) if idxs else 0) + 1

    plot_idx = get_next_plot_index(save_dir)

    # --- plot: lines per phase across metrics ---
    ax = mean_df.plot(marker='o', figsize=(12, 6), title="Mean Value Comparisons")
    ax.set_ylabel("Mean Value")
    ax.grid(True, axis='y', linestyle='--', linewidth=0.7)
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()

    save_path = os.path.join(save_dir, f"metric_meanline_{plot_idx}.png")
    plt.savefig(save_path)
    plt.show()
```

```python
def evaluate_traffic_and_buffer_metrics_boxplot(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
    save_dir=r"C:\Users\deniz\OneDrive\Masast\research_evaluation_photos"
):
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
    import os
    from pathlib import Path

    # Raw metric name -> pretty label
    metrics = {
        'receivedPacketFromUpperLayer:sum(packetBytes)': 'Received from Upper Layer (bytes)',
        'rcvdPkFromHl:sum(packetBytes)':                 'Received from HL (bytes)',
        'packetSent:sum(packetBytes)':                   'Packet Sent (bytes)',
        'packetReceived:sum(packetBytes)':               'Packet Received (bytes)',
        'sentPacketToLowerLayer:sum(packetBytes)':       'Sent to Lower Layer (bytes)',
        'receivedPacketFromLowerLayer:sum(packetBytes)': 'Received from Lower Layer (bytes)',
        'queueBitLength:max':                            'Queue Max Bit Length',
        'receivedPacketFromLowerLayer:count':            'Packets from Lower Layer (count)',
    }

    # --- helpers ---
    def load_df(path):
        return pd.read_csv(path, low_memory=False)

    def prep_scalar_subset(df, phase):
        if 'type' not in df.columns:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf = df[df['type'] == 'scalar'].copy()
        if sdf.empty or 'name' not in sdf.columns or 'value' not in sdf.columns:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf = sdf[sdf['name'].isin(metrics.keys())].copy()
        if sdf.empty:
            return pd.DataFrame(columns=['name', 'value', 'phase'])
        sdf['value'] = pd.to_numeric(sdf['value'], errors='coerce')
        sdf['phase'] = phase
        return sdf

    # --- load & prepare each dataset ---
    before   = prep_scalar_subset(load_df(before_path),   'Before')
    worst    = prep_scalar_subset(load_df(worst_path),    'After - Worst')
    moderate = prep_scalar_subset(load_df(moderate_path), 'After - Moderate')
    best     = prep_scalar_subset(load_df(best_path),     'After - Best')

    combined = pd.concat([before, worst, moderate, best], ignore_index=True)
    if combined.empty:
        print("[Info] No matching scalar metrics found in the provided files.")
        return

    # Pretty labels + consistent phase order
    combined['Metric'] = combined['name'].map(metrics)
    combined = combined.dropna(subset=['Metric', 'value'])
    combined['phase'] = pd.Categorical(
        combined['phase'],
        categories=['Before', 'After - Worst', 'After - Moderate', 'After - Best'],
        ordered=True
    )

    # Ensure directory exists and set starting index once
    Path(save_dir).mkdir(parents=True, exist_ok=True)

    def get_next_index(path, prefix):
        files = [f for f in os.listdir(path) if f.startswith(prefix) and f.endswith('.png')]
        idxs = []
        for f in files:
            try:
                idxs.append(int(f.rsplit('.', 1)[0].split('_')[-1]))
            except Exception:
                pass
        return (max(idxs) if idxs else 0) + 1

    prefix = "traffic_buffer_boxplot_"
    plot_idx = get_next_index(save_dir, prefix)

    sns.set(style="whitegrid")

    # --- one boxplot per metric across phases ---
    for pretty_label in metrics.values():
        subset = combined[(combined['Metric'] == pretty_label) & (~combined['value'].isna())]
        if subset.empty:
            print(f"[Info] Skipping '{pretty_label}': no numeric values found.")
            continue

        plt.figure(figsize=(10, 5))
        sns.boxplot(data=subset, x='phase', y='value')  # keep neutral coloring
        plt.title(f"{pretty_label}: Before vs After Variants")
        plt.xlabel("")
        plt.ylabel("Value")
        plt.tight_layout()

        save_path = os.path.join(save_dir, f"{prefix}{plot_idx}.png")
        plt.savefig(save_path)
        plt.show()
        plot_idx += 1
```

```python
def compare_metric_means_and_stds(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
    save_dir=r"C:\Users\deniz\OneDrive\Masast\research_evaluation_photos"
):
    import pandas as pd
    import matplotlib.pyplot as plt
    import os
    from pathlib import Path

    # Fixed metric order
    selected_metrics = [
        "outgoingPacketLengths:sum",
        "incomingPacketLengths:sum",
        "rlcCellThroughputUl:mean",
        "macCellThroughputUl:mean",
        "macCellThroughputDl:mean",
        "rlcCellThroughputDl:mean",
        "rxPkOk:sum(packetBytes)",
        "txPk:sum(packetBytes)",
        "receivedPacketFromUpperLayer:sum(packetBytes)",
        "sentPacketToLowerLayer:sum(packetBytes)",
        "receivedPacketFromLowerLayer:sum(packetBytes)",
        "passedUpPk:sum(packetBytes)",
        "rcvdPkFromHl:sum(packetBytes)",
        "sentPacketToUpperLayer:sum(packetBytes)",
        "packetReceived:sum(packetBytes)"
    ]

    # ---- helper to load and filter
    def load_df(path, phase):
        df = pd.read_csv(path, low_memory=False)
        df = df[df['type'] == 'scalar'].copy()
        df = df[df['name'].isin(selected_metrics)].copy()
        df['value'] = pd.to_numeric(df['value'], errors='coerce')
        df['phase'] = phase
        return df

    # ---- load all
    before   = load_df(before_path,   'Before')
    worst    = load_df(worst_path,    'After - Worst')
    moderate = load_df(moderate_path, 'After - Moderate')
    best     = load_df(best_path,     'After - Best')

    combined = pd.concat([before, worst, moderate, best], ignore_index=True)

    # ---- stats
    summary_stats = []
    for metric in selected_metrics:
        for phase, df in [('Before', before), ('After - Worst', worst),
                          ('After - Moderate', moderate), ('After - Best', best)]:
            vals = df[df['name'] == metric]['value']
            if not vals.empty:
                summary_stats.append({
                    "Metric": metric,
                    "Phase": phase,
                    "Mean": vals.mean(),
                    "Std": vals.std()
                })

    summary_df = pd.DataFrame(summary_stats)

    # Pivot for easy comparisons (only means shown in output)
    pivot_df = summary_df.pivot(index="Metric", columns="Phase", values="Mean")

    # Reindex to force fixed order
    pivot_df = pivot_df.reindex(selected_metrics)

    # ---- plot
    plt.figure(figsize=(15, 7))
    x = range(len(pivot_df))
    width = 0.2
    plt.bar([i - 1.5*width for i in x], pivot_df["Before"], width=width, label="Before")
    plt.bar([i - 0.5*width for i in x], pivot_df["After - Worst"], width=width, label="After - Worst")
    plt.bar([i + 0.5*width for i in x], pivot_df["After - Moderate"], width=width, label="After - Moderate")
    plt.bar([i + 1.5*width for i in x], pivot_df["After - Best"], width=width, label="After - Best")

    plt.xticks(x, pivot_df.index, rotation=40, ha="right")
    plt.ylabel("Mean Value")
    plt.title(f"Fixed 15 Metrics - Mean Comparison Across Phases")
    plt.legend(loc="upper right")
    plt.tight_layout()

    # ---- save
    Path(save_dir).mkdir(parents=True, exist_ok=True)
    existing_files = [f for f in os.listdir(save_dir) if f.startswith("metric_mean_comparison_") and f.endswith(".png")]
    indices = [int(f.split('_')[-1].split('.')[0]) for f in existing_files if f.split('_')[-1].split('.')[0].isdigit()]
    next_idx = max(indices, default=0) + 1
    filename = os.path.join(save_dir, f"metric_mean_comparison_{next_idx}.png")
    plt.savefig(filename)
    plt.show()

    return pivot_df
```

```python
def compare_queue_metrics(
    before_path="Simu5GBefore.csv",
    worst_path="Simu5GAfter_WorstCase.csv",
    moderate_path="Simu5GAfter_ModerateCase.csv",
    best_path="Simu5GAfter_BestCase.csv",
    metric_keywords=None,
    save_dir=r"C:\Users\deniz\OneDrive\Masast\research_evaluation_photos"
):
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import os
    from pathlib import Path
    from IPython.display import display

    # Default keywords
    if metric_keywords is None:
        metric_keywords = [
            'queueBitLength', 'queueDelay', 'queueLength', 'queueSize',
            'queueingDelay', 'droppedPackets'
        ]

    # ---- setup output dir + rolling index
    Path(save_dir).mkdir(parents=True, exist_ok=True)
    existing = [f for f in os.listdir(save_dir) if f.startswith("queue_metric_") and f.endswith(".png")]
    indices = []
    for f in existing:
        try:
            part = f.split("_")[2]  # queue_metric_{idx}_{keyword}.png
            if part.isdigit():
                indices.append(int(part))
        except Exception:
            pass
    next_index = (max(indices) if indices else 0) + 1

    # ---- helper to load and normalize one phase
    def load_phase(path, phase):
        df = pd.read_csv(path, low_memory=False)
        df = df[df.get('type') == 'scalar'].copy()
        if 'value' in df.columns:
            df['value'] = pd.to_numeric(df['value'], errors='coerce')
        else:
            df['value'] = pd.NA
        df['phase'] = phase
        return df[['name', 'value', 'phase']].dropna(subset=['name', 'value'])

    # ---- load four phases
    before   = load_phase(before_path,   'Before')
    worst    = load_phase(worst_path,    'After - Worst')
    moderate = load_phase(moderate_path, 'After - Moderate')
    best     = load_phase(best_path,     'After - Best')

    combined_all = pd.concat([before, worst, moderate, best], ignore_index=True)

    # ---- plotting style
    sns.set(style="whitegrid")

    # ---- loop over keywords and produce summary + boxplot per keyword
    for keyword in metric_keywords:
        subset = combined_all[combined_all['name'].str.contains(keyword, case=False, na=False)].copy()

        if subset.empty:
            print(f" No data found for keyword: '{keyword}'  skipping.")
            continue

        # Keep phase order consistent and print summary
        subset['phase'] = pd.Categorical(
            subset['phase'],
            categories=['Before', 'After - Worst', 'After - Moderate', 'After - Best'],
            ordered=True
        )
        print(f"\n------ Summary for Queue Metric: {keyword} ------")
        desc = subset.groupby('phase', observed=False)['value'].describe()
        display(desc)

        # Boxplot (assign hue to avoid seaborn deprecation warning)
        plt.figure(figsize=(10, 6))
        sns.boxplot(
            data=subset,
            x='phase',
            y='value',
            hue='phase',        # <- fixes the warning
            palette='Set2',
            legend=False        # hide duplicate legend
        )
        plt.title(f"Queue Metric Comparison: {keyword}")
        plt.xlabel("Phase")
        plt.ylabel("Value")
        plt.tight_layout()

        # Save with incremented index and keyword in filename
        safe_key = "".join(ch if ch.isalnum() or ch in ('-', '_') else '_' for ch in keyword)
        filename = f"queue_metric_{next_index}_{safe_key}.png"
        filepath = os.path.join(save_dir, filename)
        plt.savefig(filepath, dpi=300)
        print(f" Saved to: {filepath}")

        plt.show()
        next_index += 1
```
